```python
import numpy 
import pylab 
import random 
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
  
# defining the number of particles
n = 10000
  
#creating three arrays for containing x, y, and z timesteps 
x = numpy.zeros(n) 
y = numpy.zeros(n) 
z = numpy.zeros(n)

#empty lists to append values for histogram
X=[]
Y=[]
Z=[]
t=[]#time

Dx=15 #number of steps per second in x direction (alter based on diffusion coeffeciant)
Dy=15 #number of steps per second in y direction
Dz=15 #number of steps per second in z direction 
  
for i in range(1, n): 
    val = random.randint(1, 3) #depending on value will tell us if we are working in the x, y or z direction
    xx=0
    yy=0
    zz=0
    t.append(val) #we will only want length of the t list in the end so it doesn't matter what we append to it at this point
    for l in range(1, Dx): #work for x direction histogram
        s = random.randint(-1,1)
        xx += s 
    X.append(xx)
    for l in range(1, Dy):
        s = random.randint(-1,1)
        yy += s
    Y.append(yy)
    for l in range(1, Dz):
        s = random.randint(-1,1)
        zz += s
    Z.append(zz)
    
    #These if statement sections will let us add the random time step values we obtained for the histogram portion of the code (the section above), for the corresponding direction
    if val == 1: #add time step to x axis
        x[i] = x[i - 1] + xx
        y[i] = y[i - 1] 
        z[i] = z[i - 1]

    elif val == 2: #add time step to y axis
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + yy
        z[i] = z[i - 1]

    else: #add time step to z axis
        x[i] = x[i - 1] 
        y[i] = y[i - 1]
        z[i] = z[i - 1] + zz
        
    #setting the boundary conditions:
    if x[i] >= 200: 
        break
    if x[i] <= -200:
        break 
    if y[i] >= 200 or y[i] <= -200:
        break
    if z[i] >= 200 or z[i] <= -200:
        break

print('Time step is: ', len(t), 'seconds')

%matplotlib inline   
# plotting stuff: 

plt.figure(1)
ax1=plt.subplot(111)
ax1.hist(X)
plt.title('Distance from origin in x axis after timestep' )
plt.ylabel('frequency')
plt.xlabel('X')

plt.figure(2)
ax2=plt.subplot(111)
ax2.hist(Y, color='orange')
plt.title('Distance from origin in y axis after each timestep')
plt.ylabel('frequency')
plt.xlabel('Y')

plt.figure(3)
ax3=plt.subplot(111)
ax3.hist(Z, color='red')
plt.title('Distance from origin in z axis after each timestep')
plt.ylabel('frequency')
plt.xlabel('Z')

plt.figure(4, figsize=(10,10))
ax = plt.subplot(111, projection='3d')
plt.title("Random Walk After %.0f"%(len(t)) + " Seconds") 
ax.plot(x, y, z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
